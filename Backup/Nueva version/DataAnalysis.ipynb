```python
import pandas as pd
import math
humedad_relativa= pd.read_csv("PER_MOROCOCHA_IWEC.csv", sep= ";", encoding="latin-1")

# Definir una función para calcular la temperatura húmeda
def calcular_Tw(Tdb, Rh):
    Tw = (Tdb * math.atan(0.151977 * (Rh + 8.313659)**(1/2)) + math.atan(Tdb + Rh) - math.atan(Rh - 1.676331) + 0.00391838 * (Rh**(3/2)) * math.atan(0.023101 * Rh) - 4.686035)
    return Tw

humedad_relativa["tem_hum"]= humedad_relativa.apply(lambda row: calcular_Tw(row['*Tseca(°C)'], row['*HR(%)']), axis=1)

humedad_relativa.to_csv("temperaturahumeda.csv")
```
