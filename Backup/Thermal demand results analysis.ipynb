1. Leemos los datos de los resultados. Ordenamos, agregamos el n√∫mero de caso y juntamos todo en un solo archivo.

```python
import pandas as pd
import numpy as np

case1=pd.read_csv("DesignBuilder\Casos demanda\Results demand\case 1 - thermal demand - hourly.csv", encoding= "latin-1", sep=";")
case1=case1.drop([0], axis=0)
case1["case"]= 1

case2=pd.read_csv("DesignBuilder\Casos demanda\Results demand\case 2 - thermal demand - hourly.csv", encoding= "latin-1", sep=";")
case2=case2.drop([0], axis=0)
case2["case"]= 2

case3=pd.read_csv("DesignBuilder\Casos demanda\Results demand\case 3 - thermal demand - hourly.csv", encoding= "latin-1", sep=";")
case3=case3.drop([0], axis=0)
case3["case"]= 3

case4=pd.read_csv("DesignBuilder\Casos demanda\Results demand\case 4 - thermal demand - hourly.csv", encoding= "latin-1", sep=";")
case4=case4.drop([0], axis=0)
case4["case"]= 4

case5=pd.read_csv("DesignBuilder\Casos demanda\Results demand\case 5 - thermal demand - hourly.csv", encoding= "latin-1", sep=";")
case5=case5.drop([0], axis=0)
case5["case"]= 5

case6=pd.read_csv("DesignBuilder\Casos demanda\Results demand\case 6 - thermal demand - hourly.csv", encoding= "latin-1", sep=";")
case6=case6.drop([0], axis=0)
case6["case"]= 6

case7=pd.read_csv("DesignBuilder\Casos demanda\Results demand\case 7 - thermal demand - hourly.csv", encoding= "latin-1", sep=";")
case7=case7.drop([0], axis=0)
case7["case"]= 7


df= pd.concat([case1,case2, case3, case4, case5, case6, case7],axis=0)

df["Date/Time"]= df["Date/Time"].astype('str')

df["Date/Time"]= df["Date/Time"].replace("/","-")


df= df[["Date/Time", "Walls", "Ground Floors" , "Partitions (int)", "Roofs", "External Infiltration", "Total Cooling", "Zone Heating", "Occupancy", "Solar Gains Exterior Windows", "Air Temperature", "Radiant Temperature", "Operative Temperature", "DHW (Electricity)", "Outside Dry-Bulb Temperature","Diffuse Horizontal Solar","Direct Normal Solar", "case"]]

df.to_csv("CleanData/thermal_anual_df.csv", index= False)
```
