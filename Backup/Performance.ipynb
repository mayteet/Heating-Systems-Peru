```python
import pandas as pd
```

```python
df= pd.read_csv("DesignBuilder\Casos demanda\Results performance\Case 2\Flow rates.csv", sep=";", skiprows= 3)
df_temp= pd.read_csv("DesignBuilder\Casos demanda\Results performance\Case 2\Temperatures.csv", sep=";", skiprows= 3)
df_pwr= pd.read_csv("DesignBuilder\Casos demanda\Results performance\Case 2\Power.csv", sep=";", skiprows= 3)
```

```python
df= df.rename(columns={"Unnamed: 0":"Date", 
                   "SOLAR LOOP SUPPLY PUMP WATER OUTLET NODE":"Solar Loop", 
                   "TEMPERING VALVE WATER OUTLET NODE":"Branch Tempering valve", 
                   "AUX HW LOOP SUPPLY PUMP WATER OUTLET NODE": "Total Aux Loop",
                   "SOLAR LOOP WATER HEATER DHW INLET NODE":"Branch tank",
                   "3637 HEATED FLOOR HOT WATER INLET NODE": "One radiant floor"
                   })

df["case"]=3

df_temp= df_temp.rename(columns= {"Unnamed: 0" : "Date"})
df_pwr= df_pwr.rename(columns= {"Unnamed: 0" : "Date",
                                "AUX HW LOOP SUPPLY PUMP": "Bomba auxiliar loop",
                                "3707 HEATED FLOOR": "Suelo radiante",
                                "AUX HW LOOP WATER HEATER": "Calentador auxiliar",
                                "SOLAR LOOP SUPPLY PUMP": "Bomba solar loop"

    
    
})
```

```python


#Colector solar
df_temp_colector_1= df_temp[["Date","SOLAR COLLECTOR WATER INLET NODE"]]
df_temp_colector_1= df_temp_colector_1.rename(columns={"SOLAR COLLECTOR WATER INLET NODE":"temperatura"})
df_temp_colector_1["descripcion"]= "solar colector inlet" 
 
df_temp_colector_2= df_temp[["Date","SOLAR COLLECTOR WATER OUTLET NODE"]]
df_temp_colector_2= df_temp_colector_2.rename(columns={"SOLAR COLLECTOR WATER OUTLET NODE":"temperatura"})
df_temp_colector_2["descripcion"]= "solar colector outlet" 

df_temp_colector=pd.concat([df_temp_colector_1,df_temp_colector_2], axis= 0)
df_temp_colector["Equipo"]= "colector solar" 


#Tanque almacenamiento energÃ­a
df_temp_energytank_1= df_temp[["Date","SOLAR LOOP WATER HEATER HEATING INLET NODE"]]
df_temp_energytank_1= df_temp_energytank_1.rename(columns={"SOLAR LOOP WATER HEATER HEATING INLET NODE":"temperatura"})
df_temp_energytank_1["descripcion"]= "solar loop inlet"

df_temp_energytank_2= df_temp[["Date","SOLAR LOOP WATER HEATER HEATING OUTLET NODE"]]
df_temp_energytank_2= df_temp_energytank_2.rename(columns={"SOLAR LOOP WATER HEATER HEATING OUTLET NODE":"temperatura"})
df_temp_energytank_2["descripcion"]= "solar loop outlet"

df_temp_energytank_3= df_temp[["Date","SOLAR LOOP WATER HEATER DHW INLET NODE"]]
df_temp_energytank_3= df_temp_energytank_3.rename(columns={"SOLAR LOOP WATER HEATER DHW INLET NODE":"temperatura"})
df_temp_energytank_3["descripcion"]= "water inlet"


df_temp_energytank_4= df_temp[["Date","SOLAR LOOP WATER HEATER DHW OUTLET NODE"]]
df_temp_energytank_4= df_temp_energytank_4.rename(columns={"SOLAR LOOP WATER HEATER DHW OUTLET NODE":"temperatura"})
df_temp_energytank_4["descripcion"]= "water outlet"

df_temp_energytank_5= df_temp[["Date","SOLAR LOOP WATER HEATER"]]
df_temp_energytank_5= df_temp_energytank_5.rename(columns={"SOLAR LOOP WATER HEATER":"temperatura"})
df_temp_energytank_5["descripcion"]= "tank temperature"

df_temp_energytank=pd.concat([df_temp_energytank_1,df_temp_energytank_2,df_temp_energytank_3,df_temp_energytank_4,df_temp_energytank_5], axis= 0)
df_temp_energytank["Equipo"]= "storage tank"

#Calentador de agua auxiliar
df_aux_1= df_temp[["Date","AUX HW LOOP WATER HEATER DHW INLET NODE"]]
df_aux_1= df_aux_1.rename(columns={"AUX HW LOOP WATER HEATER DHW INLET NODE":"temperatura"})
df_aux_1["descripcion"]= "aux heater inlet"

df_aux_2= df_temp[["Date","AUX HW LOOP SUPPLY SIDE OUTLET"]]
df_aux_2= df_aux_2.rename(columns={"AUX HW LOOP SUPPLY SIDE OUTLET":"temperatura"})
df_aux_2["descripcion"]= "aux heater outlet"

df_temp_auxheater=pd.concat([df_aux_1,df_aux_2], axis= 0)
df_temp_auxheater["Equipo"]= "aux heater"

#Suelo radiante 
df_radiant_1=df_temp[["Date",'3637 HEATED FLOOR HOT WATER INLET NODE']]
df_radiant_1= df_radiant_1.rename(columns={'3637 HEATED FLOOR HOT WATER INLET NODE':"temperatura"})
df_radiant_1["descripcion"]= "radiant floor inlet"


df_radiant_2=df_temp[["Date",'3637 HEATED FLOOR HOT WATER OUTLET NODE']]
df_radiant_2= df_radiant_2.rename(columns={'3637 HEATED FLOOR HOT WATER OUTLET NODE':"temperatura"})
df_radiant_2["descripcion"]= "radiant floor outlet"

df_temp_radiant=pd.concat([df_radiant_1,df_radiant_2], axis= 0)
df_temp_radiant["Equipo"]= "radiant floor"


#Temperaturas ambiente
df_temp_amb=df_temp[["Date",'Environment']]
df_temp_amb= df_temp_amb.rename(columns={'Environment':"temperatura"})
df_temp_amb["descripcion"]= "ambient temperature"

df_temp_int=df_temp[["Date",'3637']]
df_temp_int= df_temp_int.rename(columns={'3637':"temperatura"})
df_temp_int["descripcion"]= "interior temperature"


df_temp_ambientes= pd.concat([df_temp_amb,df_temp_int])
df_temp_ambientes["Equipo"]= "in amb temperatures"

df_temperaturas=pd.concat([df_temp_colector,df_temp_energytank,df_temp_auxheater, df_temp_radiant,df_temp_ambientes], axis= 0)

```
